```python
import os
import base64
import csv

def convert_image_to_base64(image_path):
    """Convert an image to a base64 string."""
    with open(image_path, "rb") as image_file:
        encoded_string = base64.b64encode(image_file.read()).decode('utf-8')
    return encoded_string

def convert_images_in_folder(folder_path, output_csv):
    """Convert all images in a folder to Base64 and save to CSV."""
    with open(output_csv, mode='w', newline='') as file:
        writer = csv.writer(file)
        # Write header row
        writer.writerow(['Name', 'Base64 Image'])

        # Loop through all images in the folder
        for filename in os.listdir(folder_path):
            if filename.endswith('.jpg'):
                image_path = os.path.join(folder_path, filename)
                
                # Convert image to Base64
                base64_string = convert_image_to_base64(image_path)
                
                # Extract the name (e.g., 'face_1.jpg' becomes 'face_1')
                name = filename.split('.')[0]
                
                # Write to CSV
                writer.writerow([name, base64_string])
                print(f"Processed {filename}")

# Define folder and output CSV file
folder_path = 'enrolled_faces'  # Replace with your folder path containing .jpg files
output_csv = 'output.csv'  # This is where the results will be saved

# Call the function to process images
convert_images_in_folder(folder_path, output_csv)
```
