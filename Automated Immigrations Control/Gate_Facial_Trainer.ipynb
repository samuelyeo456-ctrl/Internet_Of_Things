```python
import cv2
import os
import numpy as np

dataset_path = "/Users/samuelyeo/Desktop/enrolled_faces"
print("Dataset path:", dataset_path)

# Check if the dataset folder exists
if not os.path.exists(dataset_path):
    print("Dataset folder not found. Please check the path.")
    exit()

# Initialize the face recognizer
recognizer = cv2.face.LBPHFaceRecognizer_create()

# Arrays to hold the face images and corresponding labels
faces = []
labels = []

# Function to load the images and labels
def load_images_from_folder(folder):
    for filename in os.listdir(folder):
        if filename.endswith(".jpg"):  # Only process jpg files
            img_path = os.path.join(folder, filename)
            # Read the image and convert to grayscale
            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
            if img is None:
                print(f"Warning: Unable to load image {filename}. Skipping.")
                continue
            # Extract label from filename (assuming the format is 'face_<label>_...')
            try:
                label = int(filename.split('_')[1])  # Extract label from filename (e.g., 'face_0_tom.jpg')
            except ValueError:
                print(f"Warning: Label extraction failed for {filename}. Skipping.")
                continue
            faces.append(img)
            labels.append(label)

# Load the images and labels
load_images_from_folder(dataset_path)

# Ensure there are faces and labels to train with
if len(faces) == 0 or len(labels) == 0:
    print("No faces or labels found. Exiting.")
    exit()

# Train the recognizer with the faces and labels
recognizer.train(faces, np.array(labels))

# Save the trained model to a file
recognizer.save("trainer.yml")
print("Training complete. Model saved as trainer.yml")
```
